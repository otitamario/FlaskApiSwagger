```python
# Instalar as bibliotecas necessárias (se ainda não estiverem instaladas)
!pip install pandas openpyxl ipywidgets

# Importar as bibliotecas
import pandas as pd
from google.colab import files
import io

# Passo 1: Permitir o upload do arquivo .xlsx
uploaded = files.upload()

# Passo 2: Verificar o nome do arquivo carregado
if len(uploaded) == 0:
    print("Nenhum arquivo foi carregado.")
else:
    # Obter o nome do primeiro arquivo carregado
    file_name = list(uploaded.keys())[0]

    # Verificar se o arquivo é do tipo .xlsx
    if not file_name.endswith('.xlsx'):
        print(f"O arquivo '{file_name}' não é um arquivo Excel (.xlsx).")
    else:
        # Passo 3: Ler o arquivo Excel usando pandas
        print(f"Lendo o arquivo '{file_name}'...")
        df = pd.read_excel(io.BytesIO(uploaded[file_name]), engine='openpyxl')

        # Passo 4: Exibir as primeiras linhas do DataFrame
        print("Planilha lida com sucesso!")
        print("\nPrimeiras linhas da planilha:")
        display(df.head())
```
